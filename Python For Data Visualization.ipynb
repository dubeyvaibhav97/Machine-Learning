# Matplotlib
Most used Python package for 2-D graphics

```python
import matplotlib.pyplot as plt
%matplotlib inline      
#sets backend to inline
```

Matplotlib provides a **plot()** function to directly plot anything

```python
import numpy as np
x = np.linspace(0,10,10)
y = x**2  
plt.plot(x,y)
```

We can add labels and titles to the plot

```python
plt.plot(x,y)
plt.xlabel('X Axis title')
plt.ylabel('Y Axis title')
plt.title('Plot Title')
```

Pandas also has a built-in implementation of matplotlib , i.e We can directly use the plot() function in a dataframe

Let's load a real world dataset

```python
import pandas as pd
df = pd.read_csv('Iris.csv')      #loading iris dataset
```

```python
df.set_index('Species',inplace=True)
```

```python
df.head()    #having a quick look of our dataset
```

```python
df.describe()      
```

```python
df.info()
```

Some basic plotting in this dataframe

```python
df['SepalLengthCm'].plot(kind='line')      #plots line for the 'SepalLengthCm' column in the dataset
```

We can create multiple subplots in same canvas using **plt.subplot()**

```python
plt.subplot(1,2,1)
df['SepalLengthCm'].plot(kind='line')
plt.subplot(1,2,2)
df['SepalWidthCm'].plot(kind='line')
```

## Area Plots
Commonly used when trying to compare two or more quantities

```python
df[['SepalLengthCm','SepalWidthCm']].plot(kind='area')
```

## Histogram
A way of representing the frequency distribution of a numeric dataset.

```python
df['PetalLengthCm'].plot(kind='hist')
plt.xlabel('Petal length in cm')
```

## Specialized Visualization Tools
### Pie Charts
A pie chart (or a circle chart) is a circular statistical graphic which is divided into slices to illustrate numerical proportion.

```python
Countries = ['China','India','USA','Indonesia','Brasil']
Population = [7550262101,1409517397,1339180127,324459463,263991379]

temp_data = pd.DataFrame(Population,Countries,columns=['Population'])
```

```python
# Visualizing population of top 5 countries 
temp_data['Population'].plot(kind='pie')
```

## Box Plot
A simple way of representing statistical data on a plot.

```python
df['PetalLengthCm'].plot(kind='box')
```

## Scatter Plot
 It reveals the correlation between two variables

```python
df.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
```

# Advanced Visualization Tool : Seaborn
Seaborn is a statistical plotting library.It's built on top of matplotlib.


```python
import seaborn as sns
```

Seaborn has many built-in datasets for practice. So for the rest of the notebook , we'll visualize tips dataset.

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Distribution Plot 
### Dist plot
Better version of matplotlib's histogram function

```python
sns.distplot(tips['tip'])
```

### Joint plot
It is used to compare between 2 variables

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

It has a *kind* parameter 

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='hex')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='kde')
```

### Pair Plot
pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns).

```python
sns.pairplot(tips)
```

It also takes an optional argument *hue* for categorical column

```python
sns.pairplot(tips,hue='sex')
```

## Categorical plots
### Barplot 
It's a general plot that allows us to aggregate the categorical data.

```python
sns.barplot(x='sex',y='tip',data=tips)
```

### Count Plot
It counts the number of occurrences

```python
sns.countplot(x='sex',data=tips)
```

## Matrix Plot 
### Heatmap
A heatmap is a graphical representation of data that uses a system of color-coding to represent different values. 

```python
sns.heatmap(tips.corr())     #corr() calculates the correlation of numerical data
```

## Regression Plot
Seaborn has many capabilities for regression plots.
It can plot regression line using a function called *lmplot()*

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```
